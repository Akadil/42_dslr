# Logistic Regression: Theory from Intuition to Practice

### What you will learn
- Why logistic regression is a **classification** model (not plain regression)
- How probabilities are modeled with the **sigmoid** function
- How the training objective comes from **maximum likelihood**
- How to interpret coefficients, tune thresholds, and evaluate performance

---

> Logistic regression is one of the most important baseline models in machine learning: fast, interpretable, and surprisingly strong when features are meaningful.

## 1. Intuition: Why Not Linear Regression?

Suppose we predict whether a student passes an exam: $y \in \{0,1\}$.

A linear model
$$
\hat{y} = w^T x + b
$$
can produce values below $0$ or above $1$, which cannot represent valid probabilities.

We need a function that maps any real number to $(0,1)$.
That function is the **sigmoid**:
$$
\sigma(z) = \frac{1}{1 + e^{-z}}.
$$

This lets us model
$$
P(y=1\mid x) = \sigma(w^T x + b).
$$

## 2. Model Formulation

Define the linear score:
$$
z = w^T x + b.
$$
Then
$$
P(y=1\mid x) = \sigma(z), \qquad P(y=0\mid x)=1-\sigma(z).
$$

A common classification rule is:
$$
\hat{y} =
\begin{cases}
1, & P(y=1\mid x) \ge 0.5 \\
0, & P(y=1\mid x) < 0.5
\end{cases}
$$
which is equivalent to checking whether $z \ge 0$.

### Decision Boundary
The boundary is linear in feature space:
$$
w^T x + b = 0.
$$
So logistic regression is a **linear classifier in transformed probability space**.

## 3. Log-Odds Interpretation (Most Important for Interpretability)

Starting from
$$
P(y=1\mid x)=p=\sigma(w^T x + b),
$$
we get
$$
\log\frac{p}{1-p} = w^T x + b.
$$

The left side is the **log-odds** (also called the logit).

Interpretation of a coefficient $w_j$:
- Increasing $x_j$ by 1 changes log-odds by $w_j$.
- Odds are multiplied by $e^{w_j}$.

So logistic regression is linear in **log-odds**, not directly in probability.

## 4. From Probability to Objective Function

For one sample $(x_i, y_i)$ with $y_i \in \{0,1\}$ and $p_i=\sigma(w^T x_i+b)$:
$$
P(y_i\mid x_i) = p_i^{y_i}(1-p_i)^{1-y_i}.
$$

For $m$ samples, likelihood:
$$
\mathcal{L}(w,b)=\prod_{i=1}^{m} p_i^{y_i}(1-p_i)^{1-y_i}.
$$

Take log to simplify optimization:
$$
\log \mathcal{L}(w,b)=\sum_{i=1}^{m}\left[y_i\log p_i + (1-y_i)\log(1-p_i)\right].
$$

We minimize the negative log-likelihood (Binary Cross-Entropy):
$$
J(w,b)= -\frac{1}{m}\sum_{i=1}^{m}\left[y_i\log p_i + (1-y_i)\log(1-p_i)\right].
$$

This objective is convex for logistic regression, so there is a single global minimum.

## 5. Optimization with Gradient Descent

Let $X \in \mathbb{R}^{m\times n}$, $y \in \mathbb{R}^{m}$, and $\hat{p}=\sigma(Xw+b)$.

Gradients of BCE:
$$
\frac{\partial J}{\partial w}=\frac{1}{m}X^T(\hat{p}-y),
\qquad
\frac{\partial J}{\partial b}=\frac{1}{m}\sum_{i=1}^{m}(\hat{p}_i-y_i).
$$

Update rule:
$$
w \leftarrow w - \alpha\frac{\partial J}{\partial w},
\qquad
b \leftarrow b - \alpha\frac{\partial J}{\partial b},
$$
where $\alpha$ is the learning rate.

Practical note:
- Too large $\alpha$: divergence/oscillation.
- Too small $\alpha$: slow learning.

## 6. Regularization (Controlling Overfitting)

### L2 (Ridge) regularization
$$
J_{L2}(w,b)=J(w,b)+\frac{\lambda}{2m}\|w\|_2^2
$$
- Shrinks weights smoothly toward zero.
- Usually stable and effective baseline.

### L1 (Lasso) regularization
$$
J_{L1}(w,b)=J(w,b)+\frac{\lambda}{m}\|w\|_1
$$
- Encourages sparse solutions (many coefficients exactly zero).
- Useful for feature selection.

$\lambda$ controls the bias-variance tradeoff:
- Larger $\lambda$ -> simpler model, higher bias, lower variance.
- Smaller $\lambda$ -> more flexible model, lower bias, higher variance.

## 7. Multiclass Logistic Regression

Binary logistic regression handles two classes.
For $K>2$ classes, common approaches are:

1. **One-vs-Rest (OvR)**
- Train $K$ binary classifiers.
- Pick the class with highest score/probability.

2. **Multinomial Logistic Regression (Softmax)**
$$
P(y=k\mid x)=\frac{e^{w_k^T x + b_k}}{\sum_{j=1}^{K} e^{w_j^T x + b_j}}.
$$
- Trains all classes jointly.
- Uses multiclass cross-entropy loss.

## 8. Evaluation and Threshold Tuning

A model can output excellent probabilities and still have weak class predictions if threshold choice is poor.

Default threshold is often $0.5$, but not always optimal.

### Key metrics
- Accuracy: overall correctness
- Precision: among predicted positives, how many are true positives
- Recall: among actual positives, how many are detected
- F1-score: harmonic mean of precision and recall
- ROC-AUC: ranking quality across thresholds

If false negatives are expensive, lower the threshold.
If false positives are expensive, raise the threshold.

## 9. Assumptions, Strengths, and Limitations

### Assumptions (practical view)
- Log-odds are approximately linear in features
- Observations are independent
- Features are informative and reasonably preprocessed

### Strengths
- Fast training and inference
- Strong baseline for tabular problems
- Highly interpretable coefficients

### Limitations
- Linear decision boundary (without engineered/nonlinear features)
- Sensitive to outliers and severe multicollinearity
- May underperform on highly nonlinear patterns

## 10. Practical Checklist Before Training

1. Scale numerical features (especially for gradient-based solvers).
2. Handle class imbalance (class weights, sampling, threshold tuning).
3. Watch for data leakage.
4. Validate with stratified splits.
5. Tune regularization strength and solver settings.
6. Inspect coefficient signs and magnitudes for sanity.

---

### Final Insight
Logistic regression is not "just simple". It is a principled probabilistic model with clear interpretability, convex optimization, and strong practical performance.

## Logistic Regression Cheat Sheet

### Core equations
$$
z = w^T x + b
$$
$$
\hat{p}=\sigma(z)=\frac{1}{1+e^{-z}}
$$
$$
J(w,b)= -\frac{1}{m}\sum_{i=1}^{m}\left[y_i\log \hat{p}_i +(1-y_i)\log(1-\hat{p}_i)\right]
$$
$$
\nabla_w J = \frac{1}{m}X^T(\hat{p}-y), \qquad
\frac{\partial J}{\partial b}=\frac{1}{m}\sum_{i=1}^{m}(\hat{p}_i-y_i)
$$

### Interpretability in one line
$$
\log\frac{p}{1-p}=w^T x+b
$$
A one-unit increase in $x_j$ multiplies odds by $e^{w_j}$.

### When to use
- You need a robust, interpretable baseline.
- Features are mostly tabular and relationships are not too nonlinear.
- You value calibration and probabilistic outputs.
